```python
'''1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.'''

average_median_income = housing_data['median_income'].mean()              #----> calculating the average value with the mean function
print("The average median income is",average_median_income)

plt.hist(housing_data["median_income"],bins = 20)                         #----> using matplotlib to present the data through histogram
plt.title("Histogram Plot of median_income")                              #----> providing a title to the presentation
plt.xlabel("Range of values")
plt.ylabel("Frequency distribution")                                      #----> labelling the x and y axis
plt.show()                                                                #----> presenting the histogram
```

```python
'''2. Draw an appropriate plot to see the distribution of housing_median_age and explain your observations.'''

plt.hist(housing_data["housing_median_age"],bins = 50)                    #----> using matplotlib to present the data through histogram
plt.title("Histogram Plot of housing_median_age")                         #----> providing a title to the presentation
plt.xlabel("Range of values")
plt.ylabel("Frequency Distribution")                                      #----> labelling the x and y axis
plt.show()                                                                #----> presenting the histogram
```

```python
'''3. Show with the help of visualization, how median_income and median_house_values are related?'''

sns.scatterplot(x = "median_income", y = "housing_median_age",data = housing_data)      #----> using scatter plot from seaborn to present the relation.
plt.show()                                                                              #----> presenting the scatterplot
```

```python
'''4. Create a data set by deleting the corresponding examples from the data set for which total_bedrooms are not available.'''

new_total_bedrooms = housing.dropna(subset=['total_bedrooms'])           #---->  using the dropna function to drop the null/missing values in the 'total_bedrooms' column
new_total_bedrooms
```

```python
'''5. Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.'''

mean_total_bedrooms = housing['total_bedrooms'].mean()                   #---->  calculating the mean value of the column
mean_total_bedrooms
```

```python
'''6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required'''

def get_dataset_median(dataset,column_name):                             #----> setting 2 parameters to the function- for dataset and for the column_name
  df = pd.read_excel(dataset)                                            #---->  reads the dataset
  median = df[column_name].median()                                      #----> using the median function on the column
  return median                                                          #----> return keyword that returns median value of the column


median = get_dataset_median("housing.xlsx","median_income")              #----> providing the required parameters here- for dataset and for the column_name
print(median)

```

```python
'''8. Create a data set for which the ocean_proximity is ‘Near ocean’.'''

new_data = housing_data[housing_data['ocean_proximity'] == 'NEAR OCEAN'] #---->  Here we have set the condition that the column should only contain a particular value
houses_near_ocean = pd.DataFrame(new_data)                               #---->  DataFraming the new data
houses_near_ocean
```

```python
'''9. Find the mean and median of the median income for the data set created in question 8.'''

mean_value= houses_near_ocean['median_income'].mean()                             #---->  calculating the mean value with the mean function
print("The mean value of median_income is",mean_value)

median_value= houses_near_ocean['median_income'].median()                         #---->  calculating the median value with the median function
print("The median value of median_income is",median_value)
```

```python
'''10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small.
If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.'''

'''For this question we have set multiple conditions on the column =total_bedrooms using the loc function that
 will add another column = total_bedroom_size to the dataset including the values of small,medium,large.'''

housing_data.loc[housing_data["total_bedrooms"] <= 10,"total_bedroom_size"]="Small"
#---->Since the bedroom size is less than or equal to 10 this will represent as "Small" in the total_bedroom_size column.
housing_data.loc[housing_data["total_bedrooms"] > 10,"total_bedroom_size"]="Medium"
#---->Since the bedroom size is greater tan 10 this will represent as "Medium" in the total_bedroom_size column.
housing_data.loc[housing_data["total_bedrooms"] <= 1000,"total_bedroom_size"]="Medium"
#---->Since the bedroom size is less  than or equal to 1000 this will represent as "Medium" in the total_bedroom_size column.
housing_data.loc[housing_data["total_bedrooms"] > 1000,"total_bedroom_size"]="Large"
#---->Since the bedroom size is greater than 1000 this will represent as "Large" in the total_bedroom_size column
housing_data
```
